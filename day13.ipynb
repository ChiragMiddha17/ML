```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('ai.jpg')
img2 = cv2.imread('myfile2.png')
```

```python
img.shape , img2.shape
```

```python
img = img[0:183 , 0:275 , :]
img.shape , img2.shape
```

```python
def showing(image):
    plt.imshow(image[:,:,::-1])
    plt.show()

def imshow(image):
    cv2.imshow('Preview',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
imshow(img)
imshow(img2)
```

```python
showing(img)
showing(img2)
```

```python
added = cv2.add(img,img2)
showing(added)
```

```python
subtracted = cv2.subtract(img,img2)
showing(subtracted)
```

```python
import numpy as np
```

```python
img2_1 = cv2.multiply(img2 , 2*np.ones(img2.shape).astype('uint8'))
showing(img2_1)
```

```python
img2_1 = cv2.subtract(img2 , 2*np.ones(img2.shape).astype('uint8'))
showing(img2_1)
```

```python
img2_2 = cv2.convertScaleAbs(img , alpha = 1.1 , beta = 7)
showing(img2_2)
```

```python
img2_3 = cv2.resize(img2 , (400,400) , interpolation = cv2.INTER_CUBIC)
img2_3.shape
```

```python
showing(img2_3)
```

```python
img2_4 = cv2.resize(img2 , None , fx = 2 , fy = 0.5 , interpolation = cv2.INTER_CUBIC)
img2_4.shape
```

```python
showing(img2_4)
```

```python
from skimage.transform import rotate
```

```python
img_5 = rotate(img2 , angle=45 , resize = True)
showing(img_5)
```

```python
from time import sleep
repeat = True
while repeat:
    for i in range(360):
        i = rotate(img , angle = i , resize = True)
        cv2.imshow('',i)
        key = cv2.waitKey(1)
        if key == ord('q'):
            repeat = False
            break
            sleep(0.1)
    cv2.destroyAllWindows()
```
